```python
def get_current_stock_price(symbol):
    """get the current stock price from yahoo finance by entering the symbol"""
    url = 'https://finance.yahoo.com/quote/' + symbol
    r = requests.get(url)
    text = r.text
    soup = BeautifulSoup(text, 'lxml')
    price = float(soup.find('span', {'class':'Trsdu(0.3s)', 'data-reactid':"14"}).text)
    return price
```

### Portfolio Value

```python
portfolio = defaultdict(None)
```

```python
def portfolio_change(ttype, company, amount, currency, price):
    """Buy or Sell a share"""
    if company in portfolio.keys():
        #update the amount and the purchase price
        if ttype == 'buy':
            purchase_value = price * amount
            portfolio[company]['Amount'] += amount
            portfolio[company]['Avg Purchase Price'] = (purchase_value + portfolio[company]['Position Value'])/portfolio[company]['Amount']
            portfolio[company]['Position Value'] = portfolio[company]['Amount'] * portfolio[company]['Avg Purchase Price']
        #update the amount and the purchase price
        else:
            portfolio[company]['Amount'] -= amount
            portfolio[company]['Realized Gain/Loss'] = amount * (price - portfolio[company]['Avg Purchase Price'])
            portfolio[company]['Position Value'] = portfolio[company]['Amount'] * portfolio[company]['Avg Purchase Price']
    else:
        portfolio[company] = {'Currency': currency,
                               'Amount': amount, 
                               'Avg Purchase Price': price,
                               'Position Value': amount * price}
```

```python
portfolio_change('buy', 'ABB', 100, 'CHF', 4.50)
portfolio_change('buy', 'ABB', 100, 'CHF', 5)
portfolio_change('sell', 'ABB', 100, 'CHF', 6)
portfolio_change('buy', 'NESN', 100, 'CHF', 56)
```

```python
portfolio_df = pd.DataFrame.from_dict(portfolio, orient = 'index')
portfolio_df
```
